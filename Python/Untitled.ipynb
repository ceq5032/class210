```python
tagsIwant = ['VBD', 'VBN', 'VBZ']
shortList = [word for word, tag in POStagged if tag in tagsIwant]
print(len(shortList))            
print(shortList)
```

```python
import nltk
from nltk import FreqDist, ConditionalFreqDist
from nltk.tokenize import word_tokenize
```

```python
pairs = [('text_section', word) for word in words]
```

```python
cfd = ConditionalFreqDist(pairs)
```

```python
import nltk
from nltk.corpus import words, stopwords
```

```python
nltk.download('words')
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
def content_fraction(text):
    stopwords_list = stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords_list]
    return len(content) / len(text) if len(text) > 0 else 0
```

```python
def process_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
            print(f"successfully read file: {file_path}")
except Exception as e:
    print(f"Error reading file: {e}")
    return
```

```python
try:
text_tokens = nltk.word_tokenize(text)
print(f"successfully tokenized text, numer of tokens: {len(text_tokens)}")
except Exception as e:
print(f"Error tokenizing text: {e}")
return
```

```python
unusual = content_fraction(text_tokens)
print(f"Unusual words: {unusual[:10]}...')
```

```python
fraction = content_fraction(text_tokens)
print(f"Content fraction: {fraction}")
```
